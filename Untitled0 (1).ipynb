```python
from collections import deque

# -----------------------------
# Tree Representation
# -----------------------------
tree = {
    'A': ['B', 'C'],
    'B': ['D', 'E'],
    'C': ['F', 'G'],
    'D': [],
    'E': [],
    'F': [],
    'G': []
}

# -----------------------------
# BFS (Breadth First Search)
# -----------------------------
def bfs_tree(tree, start):
    visited = []
    queue = deque([start])

    while queue:
        node = queue.popleft()
        visited.append(node)

        for neighbor in tree[node]:
            queue.append(neighbor)

    return visited

# -----------------------------
# DFS (Depth First Search)
# -----------------------------
def dfs_tree(tree, start):
    visited = []
    stack = [start]

    while stack:
        node = stack.pop()
        visited.append(node)

        for neighbor in reversed(tree[node]):
            stack.append(neighbor)

    return visited


# -----------------------------
# Output
# -----------------------------
print("TREE BFS:", bfs_tree(tree, 'A'))
print("TREE DFS:", dfs_tree(tree, 'A'))
```

```python
from collections import deque

# -----------------------------
# Graph Representation
# -----------------------------
graph = {
    'A': ['B', 'C'],
    'B': ['A', 'D'],
    'C': ['A', 'D'],
    'D': ['B', 'C']
}

# -----------------------------
# BFS (Breadth First Search)
# -----------------------------
def bfs_graph(graph, start):
    visited = set()
    queue = deque([start])
    result = []

    while queue:
        node = queue.popleft()

        if node not in visited:
            visited.add(node)
            result.append(node)

            for neighbor in graph[node]:
                queue.append(neighbor)

    return result

# -----------------------------
# DFS (Depth First Search)
# -----------------------------
def dfs_graph(graph, start):
    visited = set()
    stack = [start]
    result = []

    while stack:
        node = stack.pop()

        if node not in visited:
            visited.add(node)
            result.append(node)

            for neighbor in graph[node]:
                stack.append(neighbor)

    return result


# -----------------------------
# Output
# -----------------------------
print("GRAPH BFS:", bfs_graph(graph, 'A'))
print("GRAPH DFS:", dfs_graph(graph, 'A'))
```

```python
import random
import time
import matplotlib.pyplot as plt
import pandas as pd
from collections import deque

# -----------------------------
# BST Node
# -----------------------------
class Node:
    def __init__(self, value):
        self.value = value
        self.left = None
        self.right = None

# -----------------------------
# Insert BST
# -----------------------------
def insert(root, value):
    if root is None:
        return Node(value)

    if value < root.value:
        root.left = insert(root.left, value)
    else:
        root.right = insert(root.right, value)

    return root

# -----------------------------
# Build Tree
# -----------------------------
def build_tree(arr):
    root = None
    for num in arr:
        root = insert(root, num)
    return root

# -----------------------------
# BFS
# -----------------------------
def bfs(root, goal):
    queue = deque([root])

    while queue:
        node = queue.popleft()

        if node.value == goal:
            return True

        if node.left:
            queue.append(node.left)
        if node.right:
            queue.append(node.right)

    return False

# -----------------------------
# DFS
# -----------------------------
def dfs(root, goal):
    stack = [root]

    while stack:
        node = stack.pop()

        if node.value == goal:
            return True

        if node.right:
            stack.append(node.right)
        if node.left:
            stack.append(node.left)

    return False

# -----------------------------
# Sizes (AS REQUIRED BY PDF)
# -----------------------------
sizes = [1000, 40000, 80000, 200000, 1000000]

results = []

# -----------------------------
# MAIN LOOP
# -----------------------------
for size in sizes:
    print(f"Processing size: {size}")

    data = random.sample(range(size * 10), size)
    root = build_tree(data)

    goal = data[len(data) - 220]

    # -------------------------
    # BFS TIME (more precise)
    # -------------------------
    start = time.perf_counter()
    bfs(root, goal)
    bfs_time = time.perf_counter() - start

    # -------------------------
    # DFS TIME
    # -------------------------
    start = time.perf_counter()
    dfs(root, goal)
    dfs_time = time.perf_counter() - start

    # IMPORTANT: scale for visualization only
    results.append([
        size,
        bfs_time * 1000,
        dfs_time * 1000
    ])

# -----------------------------
# DATAFRAME
# -----------------------------
df = pd.DataFrame(results, columns=["Input Size", "BFS Time (ms)", "DFS Time (ms)"])

print("\nRESULT TABLE:")
print(df)

# -----------------------------
# PLOT (CLEAR BIG GRAPH)
# -----------------------------
plt.figure(figsize=(12,6))

x = range(len(df["Input Size"]))

plt.bar([i - 0.2 for i in x], df["BFS Time (ms)"], width=0.4, label="BFS")
plt.bar([i + 0.2 for i in x], df["DFS Time (ms)"], width=0.4, label="DFS")

plt.xticks(x, df["Input Size"])

plt.xlabel("Input Size")
plt.ylabel("Time (milliseconds)")
plt.title("BFS vs DFS Execution Time Comparison (Clear Graph)")
plt.legend()

plt.show()
```

```python
from collections import deque

# -----------------------------
# Graph Representation (Cities)
# -----------------------------
graph = {
    "Islamabad": ["Rawalpindi", "Lahore", "Peshawar"],
    "Rawalpindi": ["Islamabad", "Peshawar", "Quetta"],
    "Peshawar": ["Islamabad", "Rawalpindi", "Quetta"],
    "Lahore": ["Islamabad", "Multan", "Quetta"],
    "Multan": ["Lahore", "Karachi", "Quetta"],
    "Quetta": ["Rawalpindi", "Peshawar", "Multan", "Karachi"],
    "Karachi": ["Multan", "Quetta"]
}

# -----------------------------
# BFS (Shortest Path Function)
# -----------------------------
def bfs_shortest_path(graph, start, goal):
    visited = set()
    queue = deque([[start]])   # storing paths

    while queue:
        path = queue.popleft()
        city = path[-1]

        # Goal Check
        if city == goal:
            return path

        if city not in visited:
            visited.add(city)

            for neighbor in graph[city]:
                new_path = list(path)
                new_path.append(neighbor)
                queue.append(new_path)

    return None

# -----------------------------
# Input (Start & Goal)
# -----------------------------
start_city = "Islamabad"
goal_city = "Karachi"

# -----------------------------
# Execution
# -----------------------------
shortest_path = bfs_shortest_path(graph, start_city, goal_city)

# -----------------------------
# Output
# -----------------------------
print("Shortest Path:", shortest_path)
print("Number of Cities:", len(shortest_path))
print("Formatted Path:", " -> ".join(shortest_path))
```
